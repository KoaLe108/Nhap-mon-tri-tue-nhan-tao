**Thực hành 17 - Giải thuật Minimax cho cờ caro 5x5:**

- Thực hành này mở rộng ý tưởng từ Thực hành 16 bằng cách áp dụng cùng một thuật toán Minimax để giải bài toán cờ caro (Game XO) trên một bàn cờ lớn hơn, có kích thước 5x5.

- Sự khác biệt chính là việc mở rộng kích thước bàn cờ, điều này có thể làm tăng độ phức tạp tính toán của thuật toán Minimax do số lượng trạng thái (nước đi) có thể có tăng lên đáng kể.

- Giống như Thực hành 16, bài thực hành này cũng là một phần của môn học Trí tuệ nhân tạo và nhằm mục đích giúp học viên hiểu rõ hơn về cách thuật toán Minimax hoạt động trong các trò chơi có độ phức tạp khác nhau.

```python
from tkinter import Tk, Button, messagebox, Menu
import numpy as np

class OXGame:
    def __init__(self, size=3):
        self.root = Tk()
        self.root.title(f'Game XO {size}x{size}')
        self.size = size
        self.board = np.full((size, size), ' ')
        self.current_player = 'X'
        self.buttons = []
        self.game_over = False

        self.create_menu()
        self.create_board()

    def create_menu(self):
        menubar = Menu(self.root)
        self.root.config(menu=menubar)
        options_menu = Menu(menubar, tearoff=0)
        menubar.add_cascade(label="Options", menu=options_menu)
        options_menu.add_command(label="New Game", command=self.reset_game)

    def create_board(self):
        for i in range(self.size):
            row = []
            for j in range(self.size):
                btn = Button(self.root, text=' ', font=('Helvetica', 20),
                            height=2, width=5,
                            command=lambda i=i, j=j: self.on_click(i, j))
                btn.grid(row=i, column=j)
                row.append(btn)
            self.buttons.append(row)

    def on_click(self, i, j):
        if self.board[i][j] == ' ' and not self.game_over:
            self.board[i][j] = self.current_player
            self.buttons[i][j].config(text=self.current_player)

            if self.check_win(i, j):
                self.highlight_winning_cells()
                messagebox.showinfo("Game Over", f"Player {self.current_player} wins!")
                self.game_over = True
            elif np.all(self.board != ' '):
                messagebox.showinfo("Game Over", "It's a draw!")
                self.game_over = True
            else:
                self.current_player = 'O' if self.current_player == 'X' else 'X'

    def check_win(self, row, col):
        # Kiểm tra hàng
        if all(self.board[row][c] == self.current_player for c in range(self.size)):
            return True
        # Kiểm tra cột
        if all(self.board[r][col] == self.current_player for r in range(self.size)):
            return True
        # Kiểm tra đường chéo chính
        if row == col and all(self.board[i][i] == self.current_player for i in range(self.size)):
            return True
        # Kiểm tra đường chéo phụ
        if row + col == self.size - 1 and all(self.board[i][self.size-1-i] == self.current_player for i in range(self.size)):
            return True
        return False

    def highlight_winning_cells(self):
        for i in range(self.size):
            for j in range(self.size):
                if self.board[i][j] == self.current_player:
                    self.buttons[i][j].config(bg='#80ffaa')

    def reset_game(self):
        self.board = np.full((self.size, size), ' ')
        self.current_player = 'X'
        self.game_over = False
        for i in range(self.size):
            for j in range(self.size):
                self.buttons[i][j].config(text=' ', bg='SystemButtonFace')

    def run(self):
        self.root.mainloop()

# Sử dụng
if __name__ == "__main__":
    game = OXGame(size=4)  # Có thể thay đổi size thành 3, 4, 5...
    game.run()
```
